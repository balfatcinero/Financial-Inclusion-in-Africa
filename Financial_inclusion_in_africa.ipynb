# Financial Inclusion in Africa

## Problem statement: Predict who is most likely to have a bank account



```
# import libraries

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```
# Ingest data

url = 'https://raw.githubusercontent.com/Mikovhe/Datasets/main/FIA/data/Train_v2.csv'

data = pd.read_csv(url,index_col=0)
```

# EDA: Exploratory data analysis

```
for column in data.columns:
  plt.hist(data[column])
  plt.title("Histogram for "+ column +" distribution")
  plt.ylabel("number of people")
  plt.show()
```
